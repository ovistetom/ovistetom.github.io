```python
import torch
import torchaudio
import matplotlib.pyplot as plt
from pydub import AudioSegment
import os
```

```python
stft = torchaudio.transforms.Spectrogram(
    n_fft=400,
    win_length=400,
    hop_length=200,
    window_fn=torch.hamming_window,
    power=1,
    normalized=True,
    center=False,
)
```

```python
path_audio = r"C:\Users\toviste\Local_Documents\code\ovistetom.github.io\icassp-2026\audio"
path_images = r"C:\Users\toviste\Local_Documents\code\ovistetom.github.io\icassp-2026\images"
```

```python
plt.ioff()
for sample_name in os.listdir(path_audio):
    sample_path = os.path.join(path_audio, sample_name)
    for file_name in os.listdir(sample_path):
        file_path = os.path.join(sample_path, file_name)
        if file_name.endswith(".mp3"):
            waveform, sample_rate = torchaudio.load(file_path)
            specgram = stft(waveform[0])
            fig, axs = plt.subplots(1, 1, figsize=(12, 8))
            plt.imshow(20*torch.log10(specgram+1e-8), vmin=-60, vmax=+10, aspect='auto', origin='lower', cmap='inferno')
            axs.set_axis_off()
            axs.set_ylim(0, 150)
            fig.tight_layout()
            plt.savefig(os.path.join(path_images, sample_name, file_name.replace(".mp3", ".png")))
```
