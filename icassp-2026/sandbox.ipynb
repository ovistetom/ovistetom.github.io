```python
path_noisy_mixture = os.path.join(root, sample_id, "noisy_mixture.flac")
path_target_speech = os.path.join(root, sample_id, "target_speech.flac")
path_filtered_mu1 = os.path.join(root, sample_id, "mu1_filtered_mixture.flac")
path_filtered_mu10 = os.path.join(root, sample_id, "mu10_filtered_mixture.flac")
path_filtered_mu01 = os.path.join(root, sample_id, "mu01_filtered_mixture.flac")

noisy_mixture, sr = sf.read(path_noisy_mixture)
target_speech, _ = sf.read(path_target_speech)
filtered_mu1, _ = sf.read(path_filtered_mu1)
filtered_mu10, _ = sf.read(path_filtered_mu10)
filtered_mu01, _ = sf.read(path_filtered_mu01)

noisy_mixture = torch.from_numpy(noisy_mixture).T[0]
target_speech = torch.from_numpy(target_speech).T[0]
filtered_mu1 = torch.from_numpy(filtered_mu1)
filtered_mu10 = torch.from_numpy(filtered_mu10)
filtered_mu01 = torch.from_numpy(filtered_mu01)
```

```python
stft = torchaudio.transforms.Spectrogram()
```

```python
for waveform, name in [
    (noisy_mixture, "noisy_mixture"),
    (target_speech, "target_speech"),
    (filtered_mu1, "filtered_mu1"),
    (filtered_mu10, "filtered_mu10"),
    (filtered_mu01, "filtered_mu01"),
]:
    specgram = stft(waveform)
    specgram = 10*torch.log10(specgram + 1e-7)
    #fig, axs = plt.subplots(figsize=(4,2))
    #plt.sca(axs)
    #plt.imshow(specgram.numpy()[:100,], aspect="auto", origin="lower", cmap="magma")
    #plt.axis("off")
    #fig.tight_layout()
    #plt.savefig(f"{name}.png")
    #plt.show()

    sf.write(f"{name}.mp3", waveform, sr)
```
