```python
import os
import librosa
import soundfile as sf
import matplotlib.pyplot as plt
```

```python
path = r"C:\Users\toviste\Local_Documents\code\ovistetom.github.io\ojsp-2026\audio"
```

```python
for id in ['01474']:
    path_id = os.path.join(path, id)
    for file_name in [f for f in os.listdir(path_id) if f.endswith('.mp3')]:

        file_path = os.path.join(path_id, file_name)

        waveform, sr = librosa.load(file_path)
        mel_spec = librosa.feature.melspectrogram(y=waveform, sr=sr, fmax=8000, n_fft=2048)
        mel_spec = librosa.power_to_db(mel_spec)

        plt.figure(figsize=(6,2))
        librosa.display.specshow(mel_spec, sr=sr, x_axis='off', y_axis='off', fmax=4000, cmap='magma')
        plt.tight_layout()
        plt.savefig(file_path.replace('.mp3', '.png'), dpi=300)
        print(file_name)
        plt.show()
```
